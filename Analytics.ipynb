```python
# Aggregate by day

messages = df.loc[df.action == 'message']
messages.set_index('date', inplace=True)
messages.sort_index(ascending=True, inplace=True)
```

```python
import matplotlib.cm as cm

clean_message_df = pd.DataFrame()

for name in messages.name.unique():
    clean_message_df[name] = messages.loc[messages.name == name, 'message'].resample("D", how="count")

clean_message_df.fillna(value=0, inplace=True)

color_map = cm.get_cmap('Accent_r')
num_of_colors = len(messages.name.unique())
colors = color_map([x/float(num_of_colors) for x in range(num_of_colors)])

with plt.style.context('ggplot'):
        clean_message_df.plot(kind='area', figsize=(13,9), grid=False, color=colors, title='Stacked messages graph')
```
